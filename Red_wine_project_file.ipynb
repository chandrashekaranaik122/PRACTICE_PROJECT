# TO Remove the outliers we are using the Zscore method

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

importing the zscore from scientific python, using numpy absolute function for removing the outlairs

```python
threshold=3
print(np.where(z>3))
```

we are setting the threshold to 3 , because 99% of the data is present in the range of 3 standard deviations and 
rest data having outliers we have remove that data

```python
z.iloc[13][9]
```

we can confirm using the iloc function its showing 5.3 means outlier

```python
z.iloc[1589][3]
```

here also outlier present

```python
df1=df[(z<3).all(axis=1)]
print(df1.shape)
print(df.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu=0
maxRs=0
for i in range(1,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    RFR=RandomForestClassifier()
    RFR.fit(x_train,y_train)
    pred=RFR.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print("Best Accuracy is", maxAccu,"at random_state ",maxRs)
```

we are passing the train data through the Randomforestclassifier to find the Random state for the model..

from this we find best Random state value is 355

```python
from sklearn.model_selection import cross_val_score
```

To avoid the overfitting and underfitting of data we are using the cross validation score ...

it will helps to overcome from the above problem

```python
print('Decision Tree classifier')
scoredt=cross_val_score(dtc,x,y,cv=5)
print('Mean score is ',scoredt.mean()*100)
print('Accuracy_score is ',accuracy_score(y_test,preddtc)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,preddtc)-scoredt.mean())*100)
print('\n')

print('Support vector mission')
scoresv=cross_val_score(svc,x,y)
print('Mean score is ',scoresv.mean()*100)
print('Accuracy score is ',accuracy_score(y_test,predsvc)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,predsvc)-scoresv.mean())*100)
print('\n')

print('GradientBoostingClassifier')
scoregd=cross_val_score(gd,x,y)
print('Mean score is ',scoregd.mean()*100)
print('Accuracy score is ',accuracy_score(y_test,predgd)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,predgd)-scoregd.mean())*100)
print('\n')

print('extraTreesClassifier')
scoreet=cross_val_score(et,x,y)
print('Mean score is ',scoreet.mean()*100)
print('Accuracy score is ',accuracy_score(y_test,predet)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,predet)-scoreet.mean())*100)
print('\n')

print('Random forest')
scorerf=cross_val_score(rf,x,y)
print('Mean score is ',scorerf.mean()*100)
print('Accuracy score is ',accuracy_score(y_test,predrf)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,predrf)-scorerf.mean())*100)
print('\n')

print('adaboost')
scoread=cross_val_score(Ad,x,y)
print('Mean score is ',scoread.mean()*100)
print('Accuracy score is ',accuracy_score(y_test,predad)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,predad)-scoread.mean())*100)
print('\n')

print('Bagging')
scorebg=cross_val_score(bg,x,y)
print('Mean score is ',scorebg.mean()*100)
print('Accuracy score is ',accuracy_score(y_test,predbg)*100)
print('Difference b/w accuracy score and cross validation score is ', (accuracy_score(y_test,predbg)-scorebg.mean())*100)
```

```python
from sklearn.model_selection import GridSearchCV

parameters={'criterion':['gini','entropy'],
           'random_state':[10,50,1000],
           'max_depth':[0,10,20],
           'n_jobs':[-2,-1,1],
           'n_estimators':[50,100,200,300]}
```

using the Hyperparameter to find the best parameters to perform the model very well

```python
GCV=GridSearchCV(ExtraTreesClassifier(),parameters,cv=5)
```

by seeing the all the model performance we can find ExtratreeClassifier is best model for this dataset and we can use the thing here to perform

```python
GCV.fit(x_train,y_train)
GCV.best_params_
```

sending the x_train and y_train through the Gridsearchcv to find best parameters of the ExtratreeClassifier

```python
model=ExtraTreesClassifier(criterion='entropy',max_depth=20,random_state=50)
```

constructing the best model using the best parameter found by gridsearchcv

```python
model.fit(x_train,y_train)
pred=model.predict(x_test)
print('accuracy_score :',accuracy_score(y_test,pred)*100)
```
